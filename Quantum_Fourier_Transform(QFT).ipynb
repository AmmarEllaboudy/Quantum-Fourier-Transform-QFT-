```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
import math
```

```python
#number of qubits
n=4
qc=QuantumCircuit(n,n)
# you can prepare your quantum state here
################
# QFT starts from here
for j in range(n):
    qc.h([j])
    for i in range(1,n-j):
        qc.cu1(-math.pi/2**(i),[j+i],[j])
    qc.barrier()
qc.measure(range(n),range(n))
back=Aer.get_backend('qasm_simulator')
result=execute(qc,backend=back,shots=2048).result()
counts=result.get_counts()
plot_histogram(counts)
```
